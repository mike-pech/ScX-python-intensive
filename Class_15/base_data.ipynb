```python
import matplotlib
import plotly.express as px
import seaborn

import numpy as np
import pandas as pd

%matplotlib inline
```

```python
data = pd.read_csv("base_data_preproc.csv")
data.head()
```

```python
data = data.drop("Unnamed: 0", axis="columns")
data.head()
```

```python
%%time
data.days_employed = abs(data.days_employed)
data.head()
```

```python
data.dob_years.describe()
```

```python
fig = px.scatter(data,
                 y="total_income",
                 x=data.index,
                color="gender",
                opacity=.3)
fig.show()
```

```python
fig = px.histogram(
    data,
    x="dob_years",
    y="total_income",
    color="education_id",
    marginal="box",
)
fig.show()
```
